```python
from lib.svm import SVMClassifier

from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train, y = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=1)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
y[y==0] = -1
```

```python
np.random.seed(0)
svm = SVMClassifier(epsilon=0.01, tol=0.01, C=1000, kernel='linear')
svm.fit(X_train_scaled, y)
```

```python
fig, ax = plt.subplots()
svm.plot_decision_boundary(ax)
```

```python
svm.decision_function(np.array([[0.2, 0.2], [1.1, 1.1]]))
```
